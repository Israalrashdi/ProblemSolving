```python
# prompt: make skewness graph

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame called 'df'
# Replace 'your_column_name' with the actual column name you want to analyze
data = pd.read_csv('/content/sample_data')

# Calculate skewness
skewness = data['TotalAmount'].skew()

# Create the skewness plot
plt.figure(figsize=(8, 6))
plt.hist(data['TotalAmount'], bins=30, edgecolor='black')  # Adjust bins as needed
plt.title(f"Skewness of TotalAmount: {skewness:.2f}")
plt.xlabel('TotalAmount')
plt.ylabel('Frequency')

# Add a vertical line for the mean
plt.axvline(data['TotalAmount'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')

plt.legend()
plt.grid(True)
plt.show()
```
