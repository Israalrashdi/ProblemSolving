```python
import numpy as np
import matplotlib.pyplot as plt

# --- Linear Regression Loss: MSE ---
def mse_loss(y_true, y_pred):
    return (y_true - y_pred) ** 2

# Generate predictions
y_true_linear = 3.0  # Target value
y_preds_linear = np.linspace(0, 6, 100)
mse_values = mse_loss(y_true_linear, y_preds_linear)

# --- Logistic Regression Loss: Binary Cross-Entropy ---
def binary_cross_entropy(y_true, y_pred_prob):
    # Add small epsilon to avoid log(0)
    epsilon = 1e-15
    y_pred_prob = np.clip(y_pred_prob, epsilon, 1 - epsilon)
    return -(y_true * np.log(y_pred_prob) + (1 - y_true) * np.log(1 - y_pred_prob))

# Generate prediction probabilities
y_preds_logistic = np.linspace(0.001, 0.999, 100)

# BCE for class 1 (positive sample)
y_true_logistic_1 = 1
bce_values_1 = binary_cross_entropy(y_true_logistic_1, y_preds_logistic)

# BCE for class 0 (negative sample)
y_true_logistic_0 = 0
bce_values_0 = binary_cross_entropy(y_true_logistic_0, y_preds_logistic)

# --- Plotting ---
plt.figure(figsize=(12, 5))

# Plot MSE
plt.subplot(1, 2, 1)
plt.plot(y_preds_linear, mse_values, color='blue')
plt.title('MSE Loss (Linear Regression)')
plt.xlabel('Predicted Value')
plt.ylabel('Loss')
plt.grid(True)

# Plot BCE
plt.subplot(1, 2, 2)
plt.plot(y_preds_logistic, bce_values_1, label='True Label = 1', color='green')
plt.plot(y_preds_logistic, bce_values_0, label='True Label = 0', color='red')
plt.title('Binary Cross-Entropy (Logistic Regression)')
plt.xlabel('Predicted Probability')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# --- Binary Cross-Entropy Loss Function ---
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15  # Avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# --- Generate Predictions ---
y_preds = np.linspace(0.001, 0.999, 300)

# Loss when true label is 1
loss_when_y1 = binary_cross_entropy(1, y_preds)

# Loss when true label is 0
loss_when_y0 = binary_cross_entropy(0, y_preds)

# --- Plotting ---
fig, ax = plt.subplots(figsize=(10, 6))

# Create a color gradient
colors = cm.plasma((y_preds - y_preds.min()) / (y_preds.max() - y_preds.min()))

# Plot loss for true label = 1
ax.plot(y_preds, loss_when_y1, color='orangered', linewidth=2.5, label='True Label = 1')

# Plot loss for true label = 0
ax.plot(y_preds, loss_when_y0, color='dodgerblue', linewidth=2.5, label='True Label = 0')

# Enhancing the Plot
ax.set_title(' Binary Cross-Entropy Loss in Logistic Regression', fontsize=16, fontweight='bold')
ax.set_xlabel('Predicted Probability (ŷ)', fontsize=13)
ax.set_ylabel('Loss', fontsize=13)
ax.legend(fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.tight_layout()
plt.show()
```
